```python
# Q.11 Given the strings "Python is awesome and fun",

# find the Length of the Longest word.

def find_longest_word_length(s):
    max_length = 0
    word_length = 0
    for char in s:
        if char == " ":
            if word_length > max_length:
                max_length = word_length
            word_length = 0
        else:
            word_length += 1
    if word_length > max_length:
        max_length = word_length
    return max_length

s = "Python is awesome and fun"
print(find_longest_word_length(s))
```

```python
#Q.12 Replace all occurrences of "apple" with "orange" in the
#strings "apple pie and apple juice".

def replace_word(s, old_word, new_word):
    result = ""
    word = ""
    for char in s:
        if char == " ":  # Check for space indicating the end of a word
            if word == old_word:  # If the current word matches old_word
                result += new_word  # Add new_word to the result
            else:
                result += word  # Add the original word to the result
            result += " "  # Add the space
            word = ""  # Reset the word
        else:
            word += char  # Build the current word
    # Handle the last word
    if word == old_word:
        result += new_word
    else:
        result += word
    return result

s = "apple pie and apple juice"
print(replace_word(s, "apple", "orange"))
```

```python
# Q.13 How would you check if the strings si "Hello" and

# s2 "hello" are equal, ignoring case?

def strings_equal_ignore_case(s1, s2):
    if len(s1) != len(s2):
        return False
    for i in range(len(s1)):
        if s1[i].lower() != s2[i].lower():
            return False
    return True

s1 = "Hello"
s2 = "hello"
print(strings_equal_ignore_case(s1, s2))

```

```python
# Q.14 From the strings "abcdefgh",

# extract every second character (1.e., "bdfh").

def extract_second_characters(s):
    result = ""
    for i in range(1, len(s), 2):
        result += s[i]
    return result

s = "abcdefgh"
print(extract_second_characters(s))

```

```python
# Q.15 Given the string s = "12345", how would you check if it

# contains only numeric digits?


def is_numeric(s):
    for char in s:
        if not ('0' <= char <= '9'):
            return False
    return True

s = "12345"
print(is_numeric(s))

```

```python
# Q.16 Convert the strings "python is fun" so that the

# first letter of each word is capitalized.

def capitalize_first_letter_of_each_word(s):
    return s.title()

s = "python is fun"
print(capitalize_first_letter_of_each_word(s))
```
